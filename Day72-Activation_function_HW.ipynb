# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline


def relu(x):
    return x if x >= 0 else 0
def dev_relu(x):
    return 1 if x >= 0 else 0


x = plt.linspace(-10,10,100)
y = []
for val in range(len(x)):
    y.append(relu(x[val]))
    
plt.plot(x, y, 'b', label='linspace(-10,10,10)')
plt.title('Relu Activation Function')
# create the graph
plt.show()

y2 = []
for val in range(len(x)):
    y2.append(dev_relu(x[val]))

plt.plot(x, y2, 'b', label='linspace(-10,10,10)')
plt.title('dev_Relu Activation Function')
# create the graph
plt.show()
```
