```python
from sklearn import datasets, metrics
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
wine = datasets.load_wine()
x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.2, random_state=11)
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
print(wine.feature_names)
print("Feature importance: ")
print(clf.feature_importances_)
```

```python
criterion = ['gini', 'entropy']
splitter = ['best', 'random']
max_depth = [2,5,7]
min_samples_split= [5, 10, 20]
min_samples_leaf = [5, 10, 15]
min_weight_fraction_leaf=0.3
best_auc = 0
for a in range(len(criterion)):
    for b in range(len(splitter)):
        for c in range(len(max_depth)):
            for d in range(len(min_samples_split)):
                for e in range(len(min_samples_leaf)):
                    print(criterion[a],splitter[b],max_depth[c],min_samples_split[d],min_samples_leaf[e],min_weight_fraction_leaf)
                    clf = DecisionTreeClassifier(
                    criterion = criterion[a]
                    ,splitter = splitter[b]
                    ,max_depth = max_depth[c]
                    ,min_samples_split = min_samples_split[d]
                    ,min_samples_leaf = min_samples_leaf[e]
                    ,min_weight_fraction_leaf = min_weight_fraction_leaf)
                    clf.fit(x_train, y_train)
                    y_pred = clf.predict(x_test)
                    acc = metrics.accuracy_score(y_test, y_pred)
                    print("Acuuracy: ", acc)
                    print(wine.feature_names)
                    print("Feature importance: ")
                    print(clf.feature_importances_)
                    if best_auc < acc:
                        best_auc = acc
print(best_auc)
```

```python
boston = datasets.load_boston()
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.2, random_state=11)
clf = DecisionTreeRegressor()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
mse = metrics.mean_squared_error(y_test, y_pred)
print("MSE: ", mse)
print(wine.feature_names)
print("Feature importance: ")
print(clf.feature_importances_)

```
